# YOLO V8 Training


```python
from ultralytics import YOLO
 
model = YOLO('yolov8s.pt')

# Training.
results = model.train(
   data='files\custom_data.yaml',
   imgsz=640,
   epochs=500,
   batch=16,
   name='Emergency_yolov8s_500s'
)

print("Training Done. Find trained models in runs.")
```
